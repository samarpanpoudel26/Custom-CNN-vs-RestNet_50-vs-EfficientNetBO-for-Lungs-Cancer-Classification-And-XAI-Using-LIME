```python
import os
import tensorflow as tf
import keras
import cv2
from keras.applications import inception_v3 as inc_net
from keras.preprocessing import image
from keras.applications.imagenet_utils import decode_predictions
from skimage.io import imread
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#Loading model
model = tf.keras.models.load_model("/Users/samarpanpoudel/Desktop/Transfer Learning/saved_models/resnet1.h5")
```

```python
IMG_SIZE = 256  

def preprocess_image(img):
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = img.astype("float32") / 255.0
    return img
```

```python
def predict_fn(images):
    images = np.array([preprocess_image(img) for img in images])
    return model.predict(images)
```

```python
image_path = "/Users/samarpanpoudel/Desktop/model 1/Dataset/Negative/LUNG MRI-NC (29).png"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
from lime import lime_image
from skimage.segmentation import mark_boundaries
```

```python
explainer = lime_image.LimeImageExplainer()

explanation = explainer.explain_instance(
    image=image,
    classifier_fn=predict_fn,
    top_labels=1,
    hide_color=0,
    num_samples=1000
)
```

```python
predicted_class = np.argmax(predict_fn([image]))

temp, mask = explanation.get_image_and_mask(
    label=predicted_class,
    positive_only=True,
    num_features=10,
    hide_rest=False
)
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(mark_boundaries(temp, mask))
plt.axis("off")
plt.savefig("003.png")
plt.show()
```
